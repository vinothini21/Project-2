```python
# importing dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
#Reading the Air_quality csv file

station_data = os.path.join("../data/all_stations.csv")
station_df = pd.read_csv(station_data, skipinitialspace = True, low_memory=False)
station_df.head()
```

```python
elec_station_df = station_df[station_df["Fuel Type Code"] == "ELEC"].reset_index()
elec_station_df
```

```python
#Pulling the needed column to a dataframe
imp_station_col = elec_station_df[["State", "EV Connector Types", "City"]]
imp_station_col.head()
```

```python
#Pulling the needed column to a dataframe
imp_station_col1 = elec_station_df[["State", "City"]]
imp_station_col1.head()
```

```python
data_grouped2 = imp_station_col1.groupby(['State']).count().reset_index()
data_grouped2.head()
```

```python
data_grouped_renamed1 = data_grouped2.rename(columns= {"City":"Number of Stations"})
data_grouped_renamed1.head()
```

```python
station_csv1 = data_grouped_renamed1.to_csv(path_or_buf='/Users/vinu/Documents/Project-2/station_data_count.csv')
```
